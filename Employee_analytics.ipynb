```python

df['age'] = df['age'].fillna(df['age'].mean())
```

```python

df['city'] = df['city'].fillna(df['city'].mode)
```

```python

df['salary'] = df['salary'].fillna(df['salary'].mean())
```

```python

df['performance_score'] = df['performance_score'].fillna(df['performance_score'].mean())
```

```python
print(df.isnull().sum().sum())
```

```python
duplicates = df[df.duplicated()]
duplicates
```

```python
print(f"Duplicate rows: {df.duplicated().sum()}")
```

```python
df = df.drop_duplicates()
```

```python
print(df.duplicated().sum())
```

```python
df['gender'] = df['gender'].str.strip().str.title()
```

```python
df['department'] = df['department'].str.strip().str.title().replace({
    'It': 'Information Technology',
    'Info Tech': 'Information Technology',
    'Hr': 'Human Resources',
})
print(df['department'])
```

```python

df['status'] = df['status'].str.strip().str.title()
df['email'] = df['email'].str.strip().str.lower()
```

```python
negative_ages = df[df['age'] < 0]
print(f"Rows with negative ages: {len(negative_ages)}")
if len(negative_ages) > 0:
    print(negative_ages[['employee_id', 'age']])
    df.loc[df['age'] < 0, 'age'] = abs(df.loc[df['age'] < 0, 'age'])
```

```python
unrealistic_ages = df[(df['age'] < 18) | (df['age'] > 70)]
print(f"\nRows with unrealistic ages: {len(unrealistic_ages)}")
if len(unrealistic_ages) > 0:
    print(unrealistic_ages[['employee_id', 'age']])
    median_age = df[(df['age'] >= 18) & (df['age'] <= 70)]['age'].median()
    df.loc[(df['age'] < 18) | (df['age'] > 70), 'age'] = median_age
```

```python
negative_salaries = df[df['salary'] < 0]
print(f"\nRows with negative salaries: {len(negative_salaries)}")
if len(negative_salaries) > 0:
    print(negative_salaries[['employee_id', 'salary']])
    df.loc[df['salary'] < 0, 'salary'] = abs(df.loc[df['salary'] < 0, 'salary'])
```

```python
invalid_scores = df[(df['performance_score'] < 0) | (df['performance_score'] > 100)]
print(f"\nRows with invalid performance scores: {len(invalid_scores)}")
if len(invalid_scores) > 0:
    print(invalid_scores[['employee_id', 'performance_score']])
    df['performance_score'] = df['performance_score'].clip(0, 100)
```

```python
highest_salary = df['salary'].max()
print("highest salary : ",highest_salary)
```

```python
lowest_salary = df['salary'].min()
print("Lowest salary : ", lowest_salary)
```

```python
Q1 = df['salary'].quantile(0.25)
Q3 = df['salary'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['salary'] < lower_bound) | (df['salary'] > upper_bound)] 
outliers
```

```python
salary_above = df[df['salary']>100000]
salary_above
```

```python
avg_with_outliers = df['salary'].mean()
avg_without_outliers = df[~df.index.isin(outliers.index)]['salary'].mean()

print(f"Average salary with outliers: ${avg_with_outliers:,.2f}")
print(f"Average salary without outliers: ${avg_without_outliers:,.2f}")
```

```python
df['age_group'] = np.where(df['age'] <= 30, 'Young (18-30)',
                  np.where(df['age'] <= 45, 'Mid (31-45)',
                  'Senior (46+)'))
```

```python
print(df[['employee_id', 'name', 'age', 'age_group']].head(10))
```

```python
salary_quantiles = df['salary'].quantile([0.33, 0.67])
print(f"Salary quantiles - 33%: ${salary_quantiles.iloc[0]:,.2f}, 67%: ${salary_quantiles.iloc[1]:,.2f}")

df['salary_category'] = pd.cut(df['salary'],
                               bins=[-float('inf'), salary_quantiles.iloc[0], salary_quantiles.iloc[1], float('inf')],
                               labels=['Low', 'Medium', 'High'])
print(df[['employee_id', 'name', 'salary', 'salary_category', 'years_experience']].head(10))
```

```python
df['experience_level'] = np.where(df['years_experience'] <= 3, 'Junior',
                          np.where(df['years_experience'] <= 7, 'Mid', 'Senior'))
print(df[['employee_id', 'name', 'salary', 'salary_category', 'years_experience', 'experience_level']].head(10))
```

```python
df['bonus_eligible'] = np.where((df['performance_score'] > 85) & (df['years_experience'] > 3), 'Yes', 'No')
df[df['bonus_eligible'] == 'Yes'][['employee_id', 'name', 'performance_score', 'years_experience', 'bonus_eligible']].head(10)
```

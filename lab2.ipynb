```python
paragraph = "Detection mechanisms also also also a a serve not only to detect failures of preventive security controls which is perhaps their main function but also to provide a degree of assurance by demonstrating when the preventive controls are indeed operating and functioning as expected." 
```

```python
data = paragraph.lower().split()
srt = sorted(data)
```

```python
print(srt)
```

```python
counts = {}
for i in srt:
    if i not in counts:
        counts[i] = 1
    else:
        add = counts[i]+1
        counts[i]=add
print(counts)
    
```

```python
import os
```

```python
file = open("data.txt", "a", encoding = "utf-8")
for data in counts:
   file_data =f"{data}\t\t\t{counts[data]}\n"
   file.write(file_data)
```

```python
file = open("data.txt", "r", encoding = "utf-8")
print(file.read())
file.close()
```

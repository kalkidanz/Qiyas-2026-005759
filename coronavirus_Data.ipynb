```python
from bs4 import BeautifulSoup
import requests
```

```python
html_doc = requests.get('https://www.worldometers.info/coronavirus/')
```

```python
soup = BeautifulSoup(html_doc.text, 'html')
```

```python
# print(soup.prettify())
```

```python
table_content = soup.find('table')
```

```python
table_head = table_content.find('thead')
table_body = table_content.find('tbody')
table_row = table_body.find_all('tr')
for data in table_row[8:]:
    # print(table_row)
    # print("--------------------------------------------------------------------------------------")
    pass
```

```python
import csv
with open('output_table.csv', mode='w', encoding = 'utf-8') as file:
    writer = csv.writer(file)
    header_data = []
    for th in table_head.find_all('th'):
        header_data.append(th.text.strip())
    writer.writerow(header_data)
        
    for row in table_row[8:]:
        row_data = []
        for td in row.find_all(['td']):
            row_data.append(td.text.strip())
        writer.writerow(row_data)

print("Data successfully written to 'output_table.csv'")
```
